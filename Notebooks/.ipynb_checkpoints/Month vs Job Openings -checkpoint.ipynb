```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
```

```python
data = pd.read_csv('../datasets/amazon_jobs_dataset.csv',encoding='utf-8')
date = data['Posting_date']
date = date.str.split(' ', expand = True)
month = date[0]
freq = month.value_counts()
```

```python
x = freq.index
height = freq
for i in range(len(x)):
    print(x[i],height[i])
```

```python
plt.figure(figsize=(8,8))
plt.bar(range(len(x)), height, tick_label = x)
plt.xticks(rotation='vertical')
plt.xlabel('Month')
plt.ylabel('No of job openings')
plt.title('Month vs Job Openings')
plt.show()
```
